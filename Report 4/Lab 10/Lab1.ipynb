```python
# Let's retry plotting without an error
import matplotlib.pyplot as plt

# Data from the table
resistance = [100000, 50000, 10000, 5000, 200, 500]  # in Ohms
v_ripple = [0.6, 0.4, 0.2, 0.8, 3.4, 3.0]  # in Volts
v_dc = [4.44, 4.76, 4.86, 4.2, 1.51, 2.32]  # in Volts

# Create the plot
plt.figure(figsize=(10, 6))

# Plot V_ripple vs. Resistance
plt.plot(resistance, v_ripple, 'o-', label='V_ripple (V)', color='b')

# Plot V_dc vs. Resistance
plt.plot(resistance, v_dc, 's-', label='V_dc (V)', color='r')

# Log scale for resistance
plt.xscale('log')

# Adding labels and title
plt.xlabel('Resistance (Ohms)')
plt.ylabel('Voltage (V)')
plt.title('V_ripple and V_dc vs Resistance')
plt.grid(True)

# Add a legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.savefig('/mnt/data/voltage_vs_resistance.png')  # Saving the plot for download
plt.show()
```

```python
# Data for Voltage (V) and Current (A) from the image
voltage = [1.0, 2.0, 3.0, 4.0, 5.6, 6.0, 7.0, 8.0, 9.0, 10.0, 15.0, 19.0]  # in Volts
current = [2.6, 9.8, 19.6, 28.6, 36.1, 46.5, 55.6, 63.5, 73.1, 82.1, 125.1, 145.1]  # in mA

# Convert current to Amperes for plotting
current = [i / 1000 for i in current]  # mA to A

# Plot Voltage vs Current
plt.figure(figsize=(10, 6))
plt.plot(voltage, current, 'o-', label='I (Current)', color='b')

# Adding labels and title
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')
plt.title('I-V Characteristics of Zener Diode')
plt.grid(True)

# Add a legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.savefig('/mnt/data/iv_curve_zener_diode.png')  # Saving the plot for download
plt.show()
```
