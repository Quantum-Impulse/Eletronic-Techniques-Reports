```python
import numpy as np
import matplotlib.pyplot as plt

# Data for Part I: Gyrating Circuits
frequencies_gyrating = np.array([400, 1000, 4000, 10000, 50000, 80000, 100000, 150000, 200000, 300000, 400000])
vout_gyrating = np.array([0.160, 0.368, 1.26, 2.66, 4.48, 4.6, 4.64, 4.72, 4.72, 4.72, 4.72])
vin_gyrating = np.array([4.6, 4.6, 4.64, 4.64, 4.64, 4.68, 4.64, 4.72, 4.72, 4.72, 4.72])
phase_shift_gyrating = np.array([100, 88, 76.5, 54.8, 14.9, 7.59, 5.03, 1.62, 0, -4.46, -7.94])
gain_gyrating = vout_gyrating / vin_gyrating

# Plot Gain as a function of frequency for Gyrating Circuits
plt.figure()
plt.semilogx(frequencies_gyrating, gain_gyrating, 'o-')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (Vout/Vin)')
plt.title('Gain as a function of frequency (Gyrating Circuits)')
plt.grid(True)
plt.show()

# Plot Phase shift as a function of frequency for Gyrating Circuits
plt.figure()
plt.semilogx(frequencies_gyrating, phase_shift_gyrating, 'o-')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Shift (degrees)')
plt.title('Phase shift as a function of frequency (Gyrating Circuits)')
plt.grid(True)
plt.show()

# Data for Part II: Active Notch Filter
frequencies_notch = np.array([1000, 5000, 8000, 10000, 13000, 15000, 16000, 19000, 22000, 28000, 50000])
vout_notch = np.array([4.6, 3.68, 2.56, 1.76, 0.680, 0.200, 0.360, 1.12, 1.68, 2.6, 3.88])
vin_notch = np.array([4.6, 4.6, 4.52, 4.48, 4.48, 4.52, 4.44, 4.48, 4.48, 4.52, 4.6])
phase_shift_notch = np.array([-8.35, -37.6, -55.2, -65.9, 76.5, 106, 78.6, 73.4, 65.4, 54.3, -8.61])
gain_notch = vout_notch / vin_notch

# Plot Gain as a function of frequency for Active Notch Filter
plt.figure()
plt.semilogx(frequencies_notch, gain_notch, 'o-')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (Vout/Vin)')
plt.title('Gain as a function of frequency (Active Notch Filter)')
plt.grid(True)
plt.show()

# Plot Phase shift as a function of frequency for Active Notch Filter
plt.figure()
plt.semilogx(frequencies_notch, phase_shift_notch, 'o-')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Shift (degrees)')
plt.title('Phase shift as a function of frequency (Active Notch Filter)')
plt.grid(True)
plt.show()
```
