```python
import matplotlib.pyplot as plt
import numpy as np

# Data for FYR 1
frequency_fyr1 = np.array([400, 1000, 2500, 4000, 15000, 30000, 35000])  # Hz
vout_fyr1 = np.array([0.4, 0.52, 1.36, 3.41, 4.18, 4.56, 3.12])  # Volts
vin_fyr1 = 8.0  # Volts
gain_fyr1 = vout_fyr1 / vin_fyr1
phase_shift_fyr1 = np.array([147.0, 83.1, 79.4, 70.2, 39.0, 25.9, 19.8])  # Degrees

# Plot Gain for FYR 1
plt.figure()
plt.plot(frequency_fyr1, gain_fyr1, 'o-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (Vout/Vin)')
plt.title('Gain as a function of frequency for FYR 1')
plt.grid(True, which="both", ls="--")
plt.savefig('fyr1_gain_plot.png')
plt.show()

# Plot Phase Shift for FYR 1
plt.figure()
plt.plot(frequency_fyr1, phase_shift_fyr1, 'o-')
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Shift (Degrees)')
plt.title('Phase Shift as a function of frequency for FYR 1')
plt.grid(True, which="both", ls="--")
plt.savefig('fyr1_phase_plot.png')
plt.show()

# Data for FYR 2
frequency_fyr2 = np.array([4000, 8000, 12000, 14000, 16000, 20000, 28000, 32000, 38000, 42000])  # Hz
vout_fyr2 = np.array([5.4, 4.6, 2.4, 1.2, 1.0, 2.4, 4.6, 5.0, 5.8, 6.2])  # Volts
vin_fyr2 = 8.0  # Volts
gain_fyr2 = vout_fyr2 / vin_fyr2
phase_shift_fyr2 = np.array([-30.0, -50.9, -61.0, -70.2, -73.4, 70.2, 60.0, 55.0, 44.3, 40.0])  # Degrees

# Plot Gain for FYR 2
plt.figure()
plt.plot(frequency_fyr2, gain_fyr2, 'o-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (Vout/Vin)')
plt.title('Gain as a function of frequency for FYR 2')
plt.grid(True, which="both", ls="--")
plt.savefig('fyr2_gain_plot.png')
plt.show()

# Plot Phase Shift for FYR 2
plt.figure()
plt.plot(frequency_fyr2, phase_shift_fyr2, 'o-')
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Shift (Degrees)')
plt.title('Phase Shift as a function of frequency for FYR 2')
plt.grid(True, which="both", ls="--")
plt.savefig('fyr2_phase_plot.png')
plt.show()
```
